```python
import pickle as pkl
import b3od.scrapper as b3s
import pandas as pd
from datetime import date, timedelta



g = b3s.B3Scrapper() #.B3Scrapper()

holidays2023 = [date(2023,1,1),date(2023,2,20),date(2023,2,21),date(2023,4,7),date(2023,4,21),date(2023,5,1),date(2023,6,8),date(2023,9,7),date(2023,10,12),date(2023,11,2),date(2023,11,15),date(2023,11,20),date(2023,12,25),date(2023,12,29)]

b3file = '/content/drive/MyDrive/b3/data/b3data_.pickle'

data = {}

with open(b3file, 'rb') as f:
    data = pkl.load(f)

if not "marketdays" in data:
  data['marketdays'] = []

def daterange(start_date, end_date):
    _md = []
    for n in range(int((end_date - start_date).days+1)):
       day = start_date + timedelta(n)
       if day.weekday() < 5:
         if not day in holidays2023:
           if not day in data['marketdays']:
            _md.append(day)

    return _md

#############################################################################
## -- yyyy-mm-dd  ---- set the data range
init_dt , end_dt  = date(2023, 1, 1) , date.today() - timedelta(1)
#############################################################################

marketdays = daterange(init_dt, end_dt)

#print(g.list_services())
print(marketdays)

data['marketdays'].extend(marketdays)
```

```python
_ctr = ['TradeInformationConsolidated','LendingOpenPosition', 'LoanBalance']
for c in _ctr:
  if not isinstance(data[c], pd.DataFrame):
    data[c] = pd.DataFrame()

    
g.start()

for c in _ctr:
  if not isinstance(data[c], pd.DataFrame):
    data[c] = pd.DataFrame()

for _dt in marketdays:
  for c in _ctr:
   _df = g.download(c,_dt)
   data[c] = pd.concat([_df, data[c]],ignore_index=True)
g.stop()
```

```python
with open(b3file, 'wb') as f:
    pkl.dump(data, f)
```
